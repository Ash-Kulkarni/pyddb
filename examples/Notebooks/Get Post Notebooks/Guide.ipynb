### DesignCheck Calc

With an ID,  calc id and variable keys (which can be found on the designcheck calc page), we can enter the values by referring to the dictionary we just made.
The “designcheck_calc()” function in the “ddb_get_post” library will take these inputs and return lists of the values, units, and a source reference of the description of the calc.

If your calculation has multiple outputs, they will all be added to these lists.

```python
calc_id = 3098194

variables = {
    'ID': job_number,
    'A_Sp': p["Net internal floor area"],
    'phi_SP': p["Small power heat gain density"]
}

value, unit, source_reference = pyddb.designcheck_calc(calc_id, job_number, variables)
```

We can now create a source and post the parameter to the same asset we took the inputs from.  We can also rerun this script to update the outputs if the inputs were to change.

```python
source_type_id = "3169ba8f-d474-4ee3-8891-09817a03b196"
source_title = f"DesignCheck Calc: {calc_id}"
source_reference = source_reference

my_source = Source()
my_source.post(
    project_id = my_project.project_id,
    source_type_id = source_type_id,
    source_title = source_title,
    source_reference = source_reference)
```

```python
parameter_type_id = "570fade0-691a-43b0-83b2-2c725aacf4eb"     # find the real type id
value = value
unit_id = "8e97e911-4624-41b3-8775-4c3ca14076a9"     # find id using id

v_a_toilets.post_parameters(
    parameter_type_id=parameter_type_id,
    value=value,
    unit_id=unit_id,
    source_id=my_source.id
)
```
