```python
import sys
sys.path.append('../../Scripts')

import ddb_get_post as pyddb
from ddb_get_post import Project, Source, Asset
from ddb_jupyter import *
```

```python
pyddb.site = "sandbox"
job_number = "07869146"
project_id = pyddb.add_new_project(job_number)
```

```python
pyddb.get_asset_types()
```

```python
product_type_id = "49202b68-1ff9-489b-a8ff-9e7d890f1bf4"

products = pyddb.get_assets(project_id = project_id,
                asset_type_id = product_type_id)
```

```python
product_ids = [x["id"] for x in products]
product_ids
```

```python
# get product parameters from ddb
product_parameters = pyddb.get_parameters(asset_id = product_ids)
product_parameters
```

```python
# format i want
{
    "product_1" : {
        "volume" : 12,
        "density" : 2400},
    
    "product_2" : {
        "volume" : 12,
        "density" : 2400}
}
```

```python
# get all product parameters
product_dict = {x["parents"][0]["name"] :{x["parameter_type"]["name"] : x["revision"]["values"][0]["value"] for x in product_parameters} for x in product_parameters}
product_dict
```

```python
# add ids to assets
for product in product_parameters:
    product_dict[product["parents"][0]["name"]]["id"] = product["id"]

product_dict
```

```python
# get a specific parameter
parameter_id = "bf5a181d-ffc0-4772-b217-8c6edf66d3c3"
project_dict = {"Project" : {x["parameter_type"]["name"] : x["revision"]["values"][0]["value"] for x in pyddb.get_parameters(parameter_id = parameter_id)}}
project_dict
```

```python
pyddb.get_source_types()
```

```python
source_type_id = "3169ba8f-d474-4ee3-8891-09817a03b196"
source_title = "A4 Notebook"
source_reference = "Jupyter"
```

```python
source_id = pyddb.add_new_source(project_id, source_type_id, source_title, source_reference)
```

```python
# perform calc for each

values = []
parameter_type_ids = []
asset_ids = []
unit_ids = []
source_ids = []

for product in product_dict:

    v = float(product_dict[product]["Volume"])
    d = float(product_dict[product]["Distance travelled by product to site"])
    p = float(product_dict[product]["Density"])
    f = float(project_dict["Project"]["Transport emission carbon factor"])
    
    a4 = v*d*p*f
    
    values.append(a4)
    parameter_type_ids.append("6cbf4335-35f7-4a17-b214-d6c3aaa53685")
    asset_ids.append(product_dict[product]["id"])
    unit_ids.append("c85f4e6e-4752-476e-816a-9525d3bbc129")
    source_ids.append("source_id")
```

```python
pyddb.add_update_parameter(
    parameter_type_ids,
    project_id,
    asset_ids,
    values,
    unit_ids,
    source_ids)
```
