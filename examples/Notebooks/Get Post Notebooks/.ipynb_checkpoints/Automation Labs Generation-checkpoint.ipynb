# Imports & Authentication

```python
import sys
sys.path.append('../../Scripts')
import pandas as pd
from ddb_auth import DDBAuth #https://github.com/arup-group/ddbpy_auth
import ddb_get_post as pyddb
import ddb_jupyter
import ipysheet
import numpy
```

## Inputs

```python
pyddb.site = "dev"  # "dev", "sandbox", or "prod"
job_numbers = ['07869129',
            '07869130',
            '07869131',
            '07869133',
            '07869134',
            '07869135',
            '07869136',
            '07869138',
            '07869139',
            '07869140',
            '07869141',
            '07869145',
            '07869146',
            '07869153',
            '07869167',
            '07869168',
            '07869173',
            '07869169',
            '07869170',
            '07869171',
            '07869125']

file_name = "..\..\Templates\Automation Labs\Automation Labs Setup.xls"
sheet_name = "Data Transformation"
```

## Read Spreadsheet

```python
df = pd.read_excel(file_name, sheet_name=sheet_name)
df[['Value']] = df[['Value']].fillna(value="no value")
df[['Unit']] = df[['Unit']].fillna(value=0)
```

## Generate GUID Maps

```python
source_types, tag_types, units, parameter_types, asset_types = ddb_jupyter.generate_guid_dicts()
```

# Upload to DDB

```python
for job_number in job_numbers:
    job_number = job_number
    
    project = pyddb.Project(job_number)
    project_id = project.project_id
    
    asset_dict = ddb_jupyter.post_all_assets(df, project_id, asset_types)
    source_dict = ddb_jupyter.post_all_sources(df, project_id, source_types)
    ddb_jupyter.post_all_parameters(df, project_id, asset_dict, source_dict, parameter_types, units)
```
