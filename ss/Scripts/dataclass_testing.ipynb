```python
import ddb_dataclasses as db
import pickle
import ddb_get_post as pyddb
```

```python
def refresh_parameter_types():
    file_name = "data/parameter_types.pkl"
    data = [db.ParameterType(**x) for x in pyddb.get_parameter_types()]
    with open(file_name, "wb") as f:
        pickle.dump(data, f)

def load_parameter_types():
    with open("data/parameter_types.pkl", 'rb') as f:
        return pickle.load(f)

def refresh_asset_types():
    file_name = "data/asset_types.pkl"
    data = [db.AssetType(**x) for x in pyddb.get_asset_types()]
    with open(file_name, "wb") as f:
        pickle.dump(data, f)

def load_asset_types():
    with open("data/asset_types.pkl", 'rb') as f:
        return pickle.load(f)

def refresh_item_types():
    file_name = "data/item_types.pkl"
    data = [db.ParameterType(**x) for x in pyddb.get_item_types()]
    with open(file_name, "wb") as f:
        pickle.dump(data, f)

def load_item_types():
    with open("data/item_types.pkl", 'rb') as f:
        return pickle.load(f)

def refresh_source_types():
    file_name = "data/source_types.pkl"
    data = [db.SourceType(**x) for x in pyddb.get_source_types()]
    with open(file_name, "wb") as f:
        pickle.dump(data, f)

def load_source_types():
    with open("data/source_types.pkl", 'rb') as f:
        return pickle.load(f)

# %%
def get_parameter_type_by_name(name: str):
    return [x for x in load_parameter_types() if x.name == name][0]

def get_asset_type_by_name(name: str):
    return [x for x in load_asset_types() if x.name == name]

def get_source_type_by_name(name: str):
    return [x for x in load_source_types() if x.name == name][0]
```
