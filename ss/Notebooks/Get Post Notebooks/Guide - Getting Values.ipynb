```python
import sys
sys.path.append('../../Scripts')
import pandas as pd
from ddb_auth import DDBAuth #https://github.com/arup-group/ddbpy_auth
import ddb_get_post as pyddb
```

```python
# enter environment
pyddb.site = "sandbox"

# enter project number
job_number = "27846500"

# get project id from project number
project_id = "3286cc10-921a-42f4-92d0-2e8d0f4cd79f"#pyddb.get_project_id(job_number)

# parameter type ids for our calculation
area_type_id = "62948765-f676-4ac3-ad3a-85f145a4253a"
power_density_id = "5212537e-56bf-4c1a-b962-98e347731c79"

# add all parameter type ids to a list
parameter_type_ids = [area_type_id, power_density_id]

# asset type id for space instances
space_instance_type_id = '21ba5fbb-f079-4c37-a3b9-f36e21e5b3ac'

# get all parameters in ddb, filtered by our environment, project, asset type, and parameter types
space_instance_parameters = pyddb.get_parameters(asset_type = space_instance_type_id,
                                                 project_id = project_id,
                                                 parameter_type_id = parameter_type_ids,
                                                page_limit = 9999)

# convert json to list of tuples containing parameter names, values, space names, space instance ids
data = [(x["parents"][0]["name"], x["parameter_type"]["name"], x["revision"]["values"][0]["value"], x["parameter_type"]["id"]) for x in space_instance_parameters]

# get a list of all unique spaces
spaces = list(set([x[0] for x in data]))

# create a dictionary of all spaces and their parameters
space_dict = {space:{x[1]: x[2] for x in data if x[0] == space} for space in spaces}

# add the asset instance id to each space in the dictionary
for space in spaces:
    for x in data:
        if x[0] == space:
            space_dict[space]["id"] = x[3]

display(space_dict)

# NOTE: this dictionary will only contain space instance that have at least one of the specified parameters
#       you will want to check if the space has all of the inputs before adding it to the list,
#       possibly reporting the spaces and their missing inputs otherwise.
```
