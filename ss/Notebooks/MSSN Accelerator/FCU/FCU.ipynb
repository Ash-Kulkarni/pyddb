```python
input_df, summer_fcu, winter_fcu = fcu_calc(
    
    job_number = "12345678",
    fcu_id = "FCU 001",
    
    summer_fresh_air_temperature = 16,
    summer_fresh_air_relative_humidity = 50,
    
    winter_fresh_air_temperature = 8,
    winter_fresh_air_relative_humidity = 80, 
    
    minimum_temperature_of_cooling_coil = 8.5,
    cooling_coil_contact_factor = 0.85,
    supply_fan_temperature_rise = 0.25,
    altitude = 0,
    
    lthw_flow_temp = 80,
    lthw_return_temp = 60,
    
    chw_flow_temp = 7,
    chw_return_temp = 13,
    chw_glycol_concentration = 50,
    chw_glycol_type = "Ethylene",

    supply_air_flow_rate = 0.6,
    fresh_air_volume_flow_rate = 0.2,

    sensible_heat_gain = 6.7,
    latent_heat_gain = 4,
    room_summer_setpoint = 23,

    winter_sensible_heat_gain = -2,
    winter_latent_heat_gain = 4,
    room_winter_setpoint = 20)

limiting_air_temperature_differential = 10
```

```python
# Applying conditional formatting to supply air temperature and relative humidity
summer_fcu.style.apply(lambda x: ["background-color: red" 
                      if target - supply > limiting_air_temperature_differential
                      else ""
                      for supply, target in
                      zip(summer_fcu["Supply Air Temperature"], input_df["Room Summer Setpoint"])],
                      subset=["Supply Air Temperature"])\
            .apply(lambda x: ["background-color: red" if v < 40
                      else "background-color: orange" if v > 60
                      else ""
                      for v in summer_fcu["Room Relative Humidity"]],
                      subset=["Room Relative Humidity"])
```

```python
# Applying conditional formatting to supply air temperature and relative humidity
winter_fcu.style.apply(lambda x: ["background-color: red" 
                      if supply - target > limiting_air_temperature_differential
                      else ""
                      for supply, target in
                      zip(winter_fcu["Supply Air Temperature"], input_df["Room Winter Setpoint"])],
                      subset=["Supply Air Temperature"])\
            .apply(lambda x: ["background-color: red" if v < 40
                      else "background-color: orange" if v > 60
                      else ""
                      for v in winter_fcu["Room Relative Humidity"]],
                      subset=["Room Relative Humidity"])
```
